This code tries to create a question answer pair from a piece of text. There might be a possibility to use this in conjunction with the image captioning of notebook 2.

```python
from question_generation.pipelines import pipeline
import requests
import json
```

```python
REQUEST = json.dumps({
    'path' : {},
    'args' : {}
})
```

```python

nlp = pipeline("question-generation")
```

```python
# GET /transformer/:caption
request = json.loads(REQUEST)
caption = request['path'].get('caption', 'Talisker')

print(json.dumps(nlp(caption)))
```
